```python
from tkinter import *
import sqlite3 as sql
import datetime
from tkinter import messagebox

conn = sql.connect('GyM.db')

#creation of the base GUI window

window = Tk()
window.title("Compound Tracker")
header = Label(window, text= "WELCOME TO GyM", font=("arial",30,"bold"), fg="black").pack()

L1 = Label(window, text = "Type of Exercise", font=("arial", 18)).place(x=10,y=100)
LID= Label(window, text = "Unique ID", font=("arial",18)).place(x=10,y=150)
L2 = Label(window, text = "Day (dd)", font=("arial",18)).place(x=10,y=200)
L3 = Label(window, text = "Month (mm)", font=("arial",18)).place(x=10,y=250)
L4 = Label(window, text = "Year (yyyy)", font=("arial",18)).place(x=10,y=300)
L5 = Label(window, text = "Weight (KG)", font=("arial",18)).place(x=10,y=350)
L6 = Label(window, text = "Name", font=("arial",18)).place(x=10,y=400)

compound = ["Running", "Crossfit", "Tabata", "Power Yoga"]
droplist1 = StringVar(window)
droplist1.set('----') #Inital placeholder for field
drop1 = OptionMenu(window, droplist1, * compound) 
drop1.place(x=220,y=105)

uid = StringVar(window)
day = StringVar(window)
month = StringVar(window)
year = StringVar(window)
weight = StringVar(window)
name = StringVar(window)

uidT = Entry(window, textvariable=uid)
uidT.place(x=220,y=150)
dayT = Entry(window, textvariable=day)
dayT.place(x=220,y=200)
monthT = Entry(window, textvariable=month)
monthT.place(x=220,y=250)
yearT = Entry(window, textvariable=year)
yearT.place(x=220,y=300)
weightT = Entry(window, textvariable=weight)
weightT.place(x=220,y=350)
nameT = Entry(window, textvariable=name)
nameT.place(x=220,y=400)    

#database operations

def create_table():
    conn.execute('CREATE TABLE IF NOT EXISTS GYMDATA (UID TEXT PRIMARY KEY NOT NULL,JOINDATE DATE NOT NULL,WEIGHT REAL NOT NULL,NAME TEXT NOT NULL,TYPE CHAR NOT NULL)')

def committodb(u,d,w,n,t):
    create_table()
    conn.execute('INSERT INTO GYMDATA (UID,JOINDATE,WEIGHT,NAME,TYPE) VALUES (?,?,?,?,?)',(u,d,w,n,t))
    conn.commit()
    
def displaydb():
    uwindow = Toplevel()
    uwindow.title("Stored Database") 
    LUID = Label(uwindow, text = "UNIQUE ID", font=("arial", 12), width = 10).grid(row = 0, column = 0)
    LJD = Label(uwindow, text = "JOIN DATE", font=("arial", 12), width = 10).grid(row = 0, column = 1)
    LWt = Label(uwindow, text = "WEIGHT", font=("arial", 12), width = 10).grid(row = 0, column = 2)
    LName = Label(uwindow, text = "NAME", font=("arial", 12), width = 10).grid(row = 0, column = 3)
    LType = Label(uwindow, text = "EXERCISE TYPE", font=("arial", 12), width = 15).grid(row = 0, column = 4)
    index = 1
    cursor = conn.execute("SELECT * FROM GYMDATA")
    for row in cursor:
        Label(uwindow, text = row[0]).grid(row = index, column = 0)
        Label(uwindow, text = row[1]).grid(row = index, column = 1)
        Label(uwindow, text = row[2]).grid(row = index, column = 2)
        Label(uwindow, text = row[3]).grid(row = index, column = 3)
        Label(uwindow, text = row[4]).grid(row = index, column = 4)
        index = index + 1
    Button(uwindow, text = "OK", command = lambda: uwindow.destroy()).grid(row = index, column = 0)
        
def updatedb():
    uwindow = Toplevel()
    uwindow.title("Update Details")
    Lupdate = Label(uwindow, text = "Enter UID", font=("arial", 12)).place(x=5,y=10)
    finduid = StringVar(uwindow)
    finduidT = Entry(uwindow, textvariable = finduid).place(x=80,y=10)
    button_d = Button(uwindow, text="Delete", command = lambda: deletedb(finduid.get()) & uwindow.destroy())
    button_d.place(x=5,y=50)
    button_m = Button(uwindow, text="Update", command = lambda: newdata(finduid.get()) & uwindow.destroy())
    button_m.place(x=80,y=50)
    
def deletedb(uid):
    cur = conn.cursor()
    cur.execute("DELETE from GYMDATA where UID = ?",(uid,))
    conn.commit()
    print ("Total number of rows deleted :", conn.total_changes)
    messagebox.showinfo("Success","Selected UID has been deleted.")

def newdata(uid):
    #gui for entry of new data
    uwindow = Toplevel()
    uwindow.title("Modify Record")
    Lupdate = Label(uwindow, text = "Enter new details", font=("arial",30,"bold"), fg="black").pack()
    uL1 = Label(uwindow, text = "Type of Exercise", font=("arial", 18)).place(x=10,y=100)
    uL2 = Label(uwindow, text = "Day (dd)", font=("arial",18)).place(x=10,y=150)
    uL3 = Label(uwindow, text = "Month (mm)", font=("arial",18)).place(x=10,y=200)
    uL4 = Label(uwindow, text = "Year (yyyy)", font=("arial",18)).place(x=10,y=250)
    uL5 = Label(uwindow, text = "Weight (KG)", font=("arial",18)).place(x=10,y=300)
    uL6 = Label(uwindow, text = "Name", font=("arial",18)).place(x=10,y=350)
    compound = ["Running", "Crossfit", "Tabata", "Power Yoga"]
    udroplist1 = StringVar(uwindow)
    udroplist1.set('----') #Inital placeholder for field
    udrop1 = OptionMenu(uwindow, udroplist1, * compound) 
    udrop1.place(x=220,y=105)
    uday = StringVar(uwindow)
    umonth = StringVar(uwindow)
    uyear = StringVar(uwindow)
    uweight = StringVar(uwindow)
    uname = StringVar(uwindow)
    udayT = Entry(uwindow, textvariable=uday).place(x=220,y=150)
    umonthT = Entry(uwindow, textvariable=umonth).place(x=220,y=200)
    uyearT = Entry(uwindow, textvariable=uyear).place(x=220,y=250)
    uweightT = Entry(uwindow, textvariable=uweight).place(x=220,y=300)
    unameT = Entry(uwindow, textvariable=uname).place(x=220,y=350)    
    #hold window for user entry
    button_ok = Button(uwindow, text="OK", command = lambda: modifydb(uid,uday.get(),umonth.get(),uyear.get(),uweight.get(),uname.get(),udroplist1.get()) & uwindow.destroy())
    button_ok.place(x=10,y=400)

def modifydb(uid,dd,mm,yyyy,w,n,t):
    if len(dd) < 2:
        dd = '0' + dd
    if len(mm) < 2:
        mm = '0' + mm
    if len(yyyy) < 4:
        for i in range (0,len(yyyy)):
            yyyy = '0' + yyyy
    date = dd + mm + yyyy
    date = datetime.datetime.strptime(date, "%d%m%Y").date()
    if date < datetime.date.today():
        errormsg()
    else:
        #database modification
        cur = conn.cursor()
        cur.execute("UPDATE GYMDATA SET JOINDATE = ?, WEIGHT = ?, NAME = ?, TYPE = ? WHERE UID = ?",(date, w, n, t, uid,))
        conn.commit()
        print ("Total number of rows modified :", conn.total_changes)
        messagebox.showinfo("Success","Selected UID has been modified.")
    

#accepting the entered values and clearing screen

def clearfnt():
    droplist1.set('----')
    uid.set('')
    day.set('')
    month.set('')
    year.set('')
    weight.set('')
    name.set('')

def errormsg():
    messagebox.showwarning("Warning","Entered Date is INVALID. Please enter again")

def fieldsverify():
    u = uid.get()
    dd = day.get()
    mm = month.get()
    yyyy = year.get()
    w = weight.get()
    n = name.get()
    t = droplist1.get()
    if len(dd) < 2:
        dd = '0' + dd
    if len(mm) < 2:
        mm = '0' + mm
    if len(yyyy) < 4:
        for i in range (0,len(yyyy)):
            yyyy = '0' + yyyy
    date = dd + mm + yyyy
    date = datetime.datetime.strptime(date, "%d%m%Y").date()
    if date < datetime.date.today():
        errormsg()
    else:
        committodb(u,date,w,n,t)

#buttons for submission and opening of database

button_1 = Button(window, text="Submit", command = fieldsverify)
button_1.place(x=10,y=450)

button_2 = Button(window,text= "Clear", command = clearfnt)
button_2.place(x=100,y=450)

button_3 = Button(window,text="Open DB", command = displaydb)
button_3.place(x=10,y=500)

button_4 = Button(window,text="Update DB", command = updatedb)
button_4.place(x=100,y=500)

button_5 = Button(window, text = "Quit", command = lambda: window.destroy())
button_5.place(x=10,y=550)

window.mainloop()
```
